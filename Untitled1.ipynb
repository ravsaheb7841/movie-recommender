```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# convert this data to
# ["Action", "Adventure", "Fantasy", "Scifi"]
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head(2)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head(2)
```

```python

def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i["name"])
            counter +=1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert)
```

```python
movies.head(2)
```

```python

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies["overview"][0]
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
# apply transformation -> removing space of two words in last four column
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head(2)
```

```python
new_df = movies[["movie_id", "title", "tags"]]
```

```python
new_df.head(2)
```

```python
# convert tags data (list) into string

new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"][0]
```

```python
# convert lower case

new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

***# Vectorization***

```python
# USE OF ntlk
# tag column words -> ['loves', 'loving', 'loved']
# convert to
# love
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
# see words
cv.get_feature_names_out()
```

```python
# calculate distance from on movie to another all movie

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# it does not change data position
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

***# Main function***

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(i[0])
```

```python
recommend("Avatar")
```
